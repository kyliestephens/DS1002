```python
!pip install english_words
from english_words import get_english_words_set
words = get_english_words_set(['web2'], lower=True)
```

```python
#1:
print(type(words)) #words is in the class 'set'
wordlist = list(words) #renaming words
print(type(wordlist))
print(len(wordlist)) #finding the length of the list

```

```python
#2
print(wordlist[:10])
#I notice that all the items listed out are strings
```

```python
#3
alpha_list = sorted(wordlist) #sorting wordlist into alphabetical order
print(alpha_list[:10]) #printing first ten alphabetical words
```

```python
#4
a_words = 0 #making sure it starts the count at 0
for x in alpha_list: #x is representing each word in the list
  if x.startswith("a"):
    a_words += 1 #adds 1 the count for each word that starts with a
print(a_words)

```

```python
#5
words_with_q = 0 #starting count at 0
for y in alpha_list: #y is representing each word in the sorted list
  if "q" in y: #if q is in word "y", then it will add one to the count
    words_with_q += 1
print(words_with_q) #answer to number of words with q

words_wo_e = 0 #starting count at 0
for w in alpha_list: #w represents each word in the sorted list
  if "e" not in w: #if e is not present in "w", one added to count
    words_wo_e += 1
print(words_wo_e) #answer to number of words w/o e
fraction = words_wo_e / len(alpha_list) #words w/o e divided by total number of
#words in list gives fraction of words w/o e
print(fraction)
```

```python
#6
import numpy as np
array_1 = np.array(alpha_list)
array_2 = np.array(wordlist)
alphabet = [chr(i) for i in range(ord('a'),ord('z')+1)]
print(alphabet)
```

This made a list of all string types of each letter in the alphabet. I printed it above.

```python
#7
import matplotlib.pyplot as plt
x = alphabet
letter_counts = {} #counting the number of words starting with each letter
for w in wordlist:
    first_letter = w[0]  #first letter of word?
    letter_counts[first_letter] = letter_counts.get(first_letter, 0) + 1
alphabetical_letter_counts = dict(sorted(letter_counts.items()))
count_data = list(alphabetical_letter_counts.values())
y = count_data
plt.bar(alphabet, count_data) #(x,y)
plt.xlabel("First Letter of Word")
plt.ylabel("Number of Words Starting with Listed Letter")
plt.title("Number of Words that Start with Each Letter in Alphabetical Order")
```

```python
#8
length_sortedlist = sorted(wordlist, key=len, reverse = True)
#reverse switches order of list so that it starts with the longest word
six_longest_words = length_sortedlist[:6]
print("These are the six longest words in wordlist", six_longest_words)
```

```python
#9
wordlength_counts = {} #creating an empty dictionary

for z in wordlist:
    wordlength = len(z)
    wordlength_counts[wordlength] = wordlength_counts.get(wordlength, 0) + 1

sorted_wordlength_counts = dict(sorted(wordlength_counts.items()))
lengths_of_words = list(sorted_wordlength_counts.keys())
counts_of_lengths = list(sorted_wordlength_counts.values())

plt.bar(lengths_of_words, counts_of_lengths)
plt.xlabel("Word Length")
plt.ylabel("Number of Words")
plt.title("Histogram of Word Lengths and Counts")
```

From the histogram, you can see the most common word length is 9 letters.
