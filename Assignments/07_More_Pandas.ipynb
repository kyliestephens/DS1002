1.  Load the following dataset into Colab and perform the requested operations in code.

 Source: https://ds1002-resources.s3.amazonaws.com/data/ride_sharing.csv

```python
#loading pandas and the data"
import pandas as pd
import numpy as np
rides = pd.read_csv('https://ds1002-resources.s3.amazonaws.com/data/ride_sharing.csv')
```

```python
# show how many rows and columns are in this dataset
rows, columns = rides.shape
print("The number of rows is:", rows)
print("The number of columns is:", columns)
```

```python
#This was to allow me to see the data frame before I began my tasks.
rides
```

```python
# The first column, "duration" is seen as a string.
# Using two snippets of code, clean this column up so that
# is now an integer (int64)
rides['duration'] = rides['duration'].str.strip('minutes')
#This code stripped the word minutes from the data in duration column, which allowed me to convert it to an integer
rides['duration'] = rides['duration'].astype('int')
```

```python
# Now display the datatypes in the updated dataframe
rides.dtypes
#As you can see below, the data type for duration is now an integer, indicated by int64
```

```python
# Show the min and max values of ride durations (the shortest and longest rides)
# Hint: If 'df' is the name of your dataframe, simply specify the column name
# and then append the .min() or .max() methods.
longestride = rides['duration'].max()
shortestride = rides['duration'].min()

print("The longest car ride is:", longestride, "minutes long")
print("The shortest car ride is:", shortestride, "minutes long")
```

```python
# Create a new dataframe object and populate it with rides that are 60 minutes or less.
# Hint: Use the .query() method for the dataframe that we used in class.
newdataset = rides.query('duration <= 60')
#this code creates a new data set with ride durations less than or equal to 60 minutes indicated by <=
```

```python
# Now show both the min and max values of the duration column for your new dataset.
newlongestride = newdataset['duration'].max()
newshortestride = newdataset['duration'].min()

print("The longest car ride is:", newlongestride, "minutes long")
print("The shortest car ride is:", newshortestride, "minutes long")
#This code follows the same path as before but uses data from newdataset for its answers
```

```python
# Using the new dataframe, update it so that is ONLY has records where
# the tire_size is 26.
updateddataset = newdataset.query('tire_size == 26')
```

```python
# Count the number of records with FEMALE rider
# Finally, produce a count of all records with FEMALE riders
# using the 26-inch wheeled rental bikes for their rentals of
# one hour or less.
finalfilter = updateddataset.query('user_gender == "FEMALE"')
count = len(finalfilter) #This gives a count of the female riders in the updateddataset using the length function
#This works because updateddataset contains only rides with 26 inch tires and durations of 60 min or less
print("The number of records of female riders using 26 inch wheeled rental bikes for their rentals of one hour or less is:", count)
```

2. Find the file `more-messy-data.csv` on Canvas and upload it into Google Colab.

Create a new Pandas DataFrame from that data and perform the following steps.

```python
#This code loads the data from the file provided on Github- did this by clicking the raw button to get the proper csv link
rides2 = pd.read_csv("https://raw.githubusercontent.com/aaron-abrams-uva/DS1002-S24/main/data/more-messy-data.csv")
# Write a snippet of code that will display the number of duplicate rows exist in the DF.
rides2.duplicated()
print(rides2.duplicated().sum())
```

```python
# Next, remove all duplicate rows using Python.
rides2 = rides2.drop_duplicates()
```

```python
# For the FOUR columns involving sepal and petal length and width, update all NaN
# values with the column mean.
columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
rides2_means = rides2.fillna(rides2[columns].mean())
rides2_means
```

```python
# Find any values in the dataframe that are still missing, and delete them.
rides2_updated = rides2_means.dropna()
rides2_updated
```

```python
# How many different species are there?
species_count = rides2_updated['species'].nunique()
print("The number of different species in this data frame is:", species_count)
```

```python
#This code allowed me to see what I needed to change for cleaning up the species variable.
species_counts = rides2_updated['species'].value_counts()
species_number = len(species_counts)
#Counts the spcies using value counts function and length function
species_names = species_counts.index.tolist()
#By converting the index species_counts to a list, I am able to see the specific changes I need to make to the names.
species_names
```

```python
# Clean up the 'species' variable so that there are exactly three different values.
#This creates a dictionary to replace mispellings of the three species with the proper species name.
#I know the changes I need to make because of the list of species names above.
rides2_updated['species'] = rides2_updated['species'].replace({
    'setoso': 'setosa',
    'Setosa': 'setosa',
    'Setosan': 'setosa',
    'Virginia': 'virginica',
    'Viginica': 'virginica',
    'Versicolor': 'versicolor'
})
rides2_updated
```

```python
#To confirm the above code worked:
species_count2 = rides2_updated['species'].nunique()
print("The number of different species in this data frame is:", species_count2)
```

The `re` package has various functions related to regular expressions, including for example `match()`, which takes two arguments, first the regular expression and second the test string, and tests to see whether the *beginning* of the `test` string matches the pattern `exp`.

Note that `re.match(exp,test)` does not return a boolean variable, as you might expect (in other words, a match doesn't return `True` and a non-match `False`).  The type of object you get is called `re.Match`, which is a data type special to the `re` package.  (The reason for this is that the object contains more information than just whether there is a match or not.) However, you can turn it into a boolean using `bool()` if you want.
